# Módulo 1 - Operações Matemáticas na Programação

Neste material, vamos aprender como realizar operações matemáticas em Python, o que é essencial para qualquer tipo de programação.

## Soma e Subtração em Python

Os operadores de soma (+) e subtração (-) funcionam da mesma forma que na matemática comum.

- `+` → soma valores  
- `-` → subtrai valores  

Essas operações podem ser feitas com números inteiros ou decimais, e seus resultados podem ser armazenados em variáveis.

Vamos ver alguns exemplos:


```python
# Demonstração prática de soma e subtração
a = 12
b = 7

print("Exemplo 1:")
print(f"{a} + {b} = {a + b}")

print("\nExemplo 2:")
print(f"{a} - {b} = {a - b}")

# Também podemos armazenar o resultado:
resultado = a + b
print(f"\nResultado armazenado em variável: resultado = {resultado}")
```

## Números Inteiros e Decimais

Python trabalha com dois tipos principais de números:

- `int` (inteiro) → números sem parte decimal (`-2`, `0`, `10`)  
- `float` (decimal) → números com ponto decimal (`2.5`, `-3.14`)

Atenção:  
Na matemática escolar usamos vírgula (,) para separar casas decimais (ex: 3,5).  
Mas em Python usamos o ponto (.).  

Não é necessário declarar o tipo: o Python reconhece automaticamente pelo valor.  
Na matemática, `int` representa os números inteiros (ℤ) e `float` os decimais (ℝ).

A função `type()` permite verificar qual é o tipo de dado armazenado em uma variável.

```python
a = 10       # número inteiro (int)
b = 2.5      # número decimal (float)

print("a =", a, "| tipo:", type(a))
print("b =", b, "| tipo:", type(b))

# Operações matemáticas
print("\nSoma:", a + b)
print("Subtração:", a - b)
print("Multiplicação:", a * b)
print("Divisão:", a / 3)
```

# Módulo 3 - Delimitadores de Expressão (Parênteses)

Neste módulo, vamos entender como os **parênteses `( )`** funcionam em Python.

## Prioridade na Matemática
Assim como na escola, o que está dentro dos parênteses é resolvido primeiro. Eles servem para alterar a ordem natural das contas.

## Atenção: Diferença Importante da Matemática
Na matemática escrita, é comum omitirmos o sinal de multiplicação antes de um parêntese, escrevendo `2(3)` para dizer `2 vezes 3`.

⚠️ **Em Python, isso gera erro!**
O Python entende `nome()` como uma **chamada de função**. Se você escrever `2(10)`, ele tentará executar o número 2 como se fosse um comando, gerando um `TypeError`.

**Forma Correta:** Você deve sempre explicitar o operador `*`.
- Errado: `5(2 + 2)`
- Correto: `5 * (2 + 2)`


```python
print("Demonstração: Uso de Parênteses")

# Sem parênteses (multiplicação acontece primeiro)
resultado1 = 10 + 5 * 2
print(f"10 + 5 * 2 = {resultado1} (O Python multiplica 5*2 primeiro)")

# Com parênteses (soma acontece primeiro)
resultado2 = (10 + 5) * 2
print(f"(10 + 5) * 2 = {resultado2} (O Python soma 10+5 primeiro)")

print("\nDemonstração do ERRO ao omitir o sinal de multiplicação:")
try:
    # Tentando fazer do jeito matemático '2(5)' sem o '*'
    eval("2(5)")
except Exception as e:
    print(f"Erro gerado: {e} -> O Python acha que '2' é uma função!")
```

# Módulo 4 - Ordem de Precedência dos Operadores

Quando uma expressão matemática tem várias operações e **não** tem parênteses, o Python segue uma ordem rígida de prioridade, assim como na matemática.

## A Regra
1. **Multiplicação (`*`) e Divisão (`/`)** são resolvidas PRIMEIRO.
2. **Soma (`+`) e Subtração (`-`)** são resolvidas DEPOIS.

Isso significa que na conta `4 + 2 * 3`, o Python primeiro fará `2 * 3` (que dá 6) e só depois somará com 4 (totalizando 10).


```python
print("Demonstração: Precedência de Operadores")

conta_teste = 4 + 2 * 3

print(f"Expressão: 4 + 2 * 3")
print(f"Se fosse na ordem de leitura (errado): 4+2=6, depois 6*3 = 18")
print(f"Como o Python faz (correto): 2*3=6, depois 4+6 = {conta_teste}")

print("\nOutro exemplo: 20 - 10 / 2")
print(f"Resultado: {20 - 10 / 2} (O Python divide 10/2 primeiro)")
```
